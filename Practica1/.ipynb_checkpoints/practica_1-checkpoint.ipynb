### 3. Python

Crear en python las siguientes celdas:

- Definir una clase "Calculadora" con métodos "suma" y "multiplicación" que tomen dos parámetros y regresen el resultado de la operación
- Definir una clase "CalculadoraMultiple" con métodos "suma" y "multiplicación" que tomen cualquier cantidad de parámetros y regresen el resultado de la operación 
- Instanciar ambas clases en una nueva celda y comprobar su funcionamiento

```python
class Calculadora:
    """Calculadora simple: operaciones con exactamente dos parámetros."""
    def suma(self, a, b):
        """Devuelve a + b"""
        return a + b

    def multiplicacion(self, a, b):
        """Devuelve a * b"""
        return a * b
```

```python
class CalculadoraMultiple:
    """Calculadora que acepta cualquier cantidad de parámetros."""
    def suma(self, *args):
        """
        Devuelve la suma de todos los argumentos.
        Si no se pasa ningún argumento, lanza ValueError.
        """
        if len(args) == 0:
            raise ValueError("Debe pasar al menos un número a suma()")
        total = 0
        for x in args:
            total += x
        return total

    def multiplicacion(self, *args):
        """
        Devuelve el producto de todos los argumentos.
        Si no se pasa ningún argumento, lanza ValueError.
        """
        if len(args) == 0:
            raise ValueError("Debe pasar al menos un número a multiplicacion()")
        producto = 1
        for x in args:
            producto *= x
        return producto
```

```python
# --- Instanciar y probar ---
print("=== Pruebas Calculadora (2 parámetros) ===")
c = Calculadora()
print("Suma 3 + 5 =", c.suma(3, 5))
print("Multiplicación 3 * 5 =", c.multiplicacion(3, 5))

print("\n=== Pruebas CalculadoraMultiple (n parámetros) ===")
cm = CalculadoraMultiple()
print("Suma 1,2,3,4 =", cm.suma(1, 2, 3, 4))
print("Multiplicación 2,3,4 =", cm.multiplicacion(2, 3, 4))

# Mostrar manejo de errores (opcional)
try:
    cm.suma()  # sin argumentos -> error
except ValueError as e:
    print("\nError esperado en suma():", e)

try:
    cm.multiplicacion()  # sin argumentos -> error
except ValueError as e:
    print("Error esperado en multiplicacion():", e)
```
